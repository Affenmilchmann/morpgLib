```python
from src.morphLib import MorphAn, token_dispersion

m_an = MorphAn()
```

```python
#m_an.load('results/master&margarita.json')
m_an.load('results/ViM.json')
```

```python
target = [
    "дуб",
    "балконский андрей"
]
```

```python
#%%timeit -r 7
out = token_dispersion(
    m_an.norm_forms, 
    target, 
    progress_bar=True, 
    do_token_normalize=False, 
    alpha=0.1, 
    squeeze=3, 
    row_size=0.8
)
```

```python
from nltk.text import Text

text = Text(m_an.norm_forms)
text.dispersion_plot([
    "родион",
    "Родион",
    "раскольников",
    "родион раскольников",
    "Алена Ивановна старуха-процентщица",
    "Дуня",
]);
```
