```python
target = [
    "и",
    "с",
    "а"
]
```

```python
#%%timeit -r 7
out = token_dispersion(
    m_an.norm_forms, 
    target, 
    progress_bar=True, 
    do_token_normalize=False, 
    alpha=0.01, 
    squeeze=2, 
    row_size=0.3
)
```

```python
from nltk.text import Text

text = Text(m_an.norm_forms)
text.dispersion_plot([
    "родион раскольников",
    "старуха-процентщица Алёна Ивановна",
    "соня мармеладова",
    "Разумихин Дмитрий Прокофьевич Вразумихин",
    "Топор",
    "беременная"
]);
```
